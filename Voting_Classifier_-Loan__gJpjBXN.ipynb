```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
loan = pd.read_csv("train_ctrUa4K.csv")
loan.head()
```

```python
loan["Married"] =  loan["Married"].fillna('Yes')
loan["Self_Employed"] = loan["Self_Employed"].fillna('No')
loan["LoanAmount"] = loan["LoanAmount"].fillna(loan["LoanAmount"].mean())
loan["Loan_Amount_Term"] = loan["Loan_Amount_Term"].fillna(360)
loan["Credit_History"] = loan["Credit_History"].fillna(1)
loan["Gender"] = loan["Gender"].fillna('Male')
```

```python
from sklearn import preprocessing
l=["Gender","Married","Education","Self_Employed",]
for i in l:
    le = preprocessing.LabelEncoder()
    le.fit(loan[i])
    loan[i]= le.transform(loan[i])
```

```python
loan = pd.get_dummies(loan, columns=['Dependents', 'Property_Area'])
```

```python
X = loan.drop(['Loan_ID', 'Loan_Status'], axis=1)
y = loan['Loan_Status']
X.head()
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8,test_size=0.2, random_state=67) 
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
```

```python
from sklearn.ensemble import VotingClassifier
voting_clf_H = VotingClassifier(estimators=[('GNB', GaussianNB()), ('RF', RandomForestClassifier()), ('DT', DecisionTreeClassifier()),('BNB', BernoulliNB())], voting='hard')
voting_clf_H.fit(X_train, y_train)
pred =voting_clf_H.predict(X_test) 
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy: \n",accuracy_score(y_test, pred))
```
